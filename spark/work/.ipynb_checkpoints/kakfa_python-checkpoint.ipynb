```python
from kafka import KafkaProducer
```

```python
producer = KafkaProducer(bootstrap_servers='broker:9092')
```

```python
producer.send('topico_teste', b'teste teste')
```

```python
from kafka import KafkaConsumer
```

```python
consumer = KafkaConsumer('topico_teste',bootstrap_servers='broker:9092')
```

```python
for msg in consumer:
    print (msg)
```
